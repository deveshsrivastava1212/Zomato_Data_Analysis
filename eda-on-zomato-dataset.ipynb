```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
# Reading CSV file
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'], axis = 1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
df['rate'].unique()
```

```python
# Remove "New", "nan", "-" from the rate columm
def cleanRate(value):
    if(value == "NEW" or value == "-"):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(cleanRate)
df['rate'].unique()
```

```python
# handle the null values
df['rate'].isnull().sum()
```

```python
# Replace all the null value in Rate Column with Mean
df['rate'].replace(np.nan, df['rate'].mean(), inplace=True)
# df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# Drop rest Row having null values in rest of the columns
df.dropna(inplace=True)
df.info()
```

```python
df.rename(columns={
    'approx_cost(for two people)':'Cost2People',
    'listed_in(type)':'Type',
    'listed_in(city)':'City'}, inplace=True)

df.columns
```

```python
df.head()
```

```python
# Listed City and Location both are there. Let's keep only one in the dataset
df = df.drop(['City'], axis=1)
```

```python
df['Cost2People'].unique()
```

```python
def handleComma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2People'] = df['Cost2People'].apply(handleComma)
df['Cost2People'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending=True)
rest_type
```

```python
rest_type_under1000 = rest_type[rest_type<1000]
rest_type_under1000
```

```python
# Making Rest types under1000 in frequency as others
def handleRest(value):
    if(value in rest_type_under1000):
        return 'others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(handleRest)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
# Make cluster of all those location who have less frequent less than 300
location = df['location'].value_counts(ascending=False)
location
```

```python
def handleLocation(value):
    if(value in location[location<300]):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handleLocation)
df['location'].value_counts()
        
```

```python
df['cuisines'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts()
def handleCuisines(value):
    if(value in cuisines[cuisines<100]):
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handleCuisines)
df['cuisines'].value_counts()
```
